```python
from bs4 import BeautifulSoup
import urllib
```

```python
r = urllib.request.urlopen('https://www.cs.purdue.edu').read()
soup = BeautifulSoup(r, 'html.parser')
```

```python
# print out results
print(soup.prettify())
```

```python
def safeOpenParsePage(targetUrl):
    try:
        tmpurl = urllib.request.urlopen(targetUrl)
        tmpR = tmpurl.read()
        tmpSoup = BeautifulSoup(tmpR, 'html.parser')
        return tmpSoup
    except urllib.error.HTTPError as e:
        print(e)
        return None

soupNYT = safeOpenParsePage('https://www.nytimes.com')
if soupNYT is not None:
    print(soupNYT.prettify())
```

```python
# find all links
links = soup.find_all('a')

# get text associated with links
for link in links:
    linkText = link.get_text().strip()
    print(linkText)
```

```python
# get urls
for link in links:
    linkUrl = link["href"]
    # filter for html pages with absolute path
    if 'http' in linkUrl and '.html' in linkUrl:
        print(linkUrl)
```

```python
# find all paragraph text without embedded links
paragraphs = soup.find_all('p')
for p in paragraphs:
    if not p.find('a'):
        pText = p.get_text()
        print(pText)        
```

```python
# count words in paragraphs on the page
word_dict = {}
for p in paragraphs:
    pText = p.get_text()
    for word in pText.split():
        if word in word_dict.keys():
            word_dict[word] = word_dict[word] + 1
        else:
            word_dict[word] = 1
            
print(word_dict)
```

```python
def printTopWordsinDict(dict, k):
    for key in dict.keys():
        if dict[key] > k:
            print(key + " " + str(dict[key]))      

# print out words that occur more than three times
printTopWordsinDict(word_dict, 3)
```

```python
def addPageWordCountsToDict(dict, soupForPage):
    paragraphs = soupForPage.find_all('p')
    for p in paragraphs:
        pText = p.get_text()
        for word in pText.split():
            if word in dict.keys():
                dict[word] = dict[word] + 1
            else:
                dict[word] = 1
    return dict

# add word counts on all linked pages
for link in links:
    linkUrl = link["href"]
    # filter for html pages with absolute path
    if 'http' in linkUrl and '.html' in linkUrl:
        print(linkUrl)
        tmpSoup = safeOpenParsePage(linkUrl)
        if tmpSoup is None:
            continue
        else:
            addPageWordCountsToDict(word_dict, tmpSoup)

# print out words that occur more than 20 times
printTopWordsinDict(word_dict, 20)
```

```python
# putting it all together
targetPage = 'https://en.wikipedia.org/wiki/Purdue_University'
soupPurdue = safeOpenParsePage(targetPage)
if soupPurdue is not None:
    # count words on in paragraphs on the page
    word_dict = {}
    addPageWordCountsToDict(word_dict, soupPurdue)
    
    # find all links
    links = soupPurdue.find_all('a', href=True)
    # add word counts on first ten linked pages
    counter = 0
    for link in links:
        linkUrl = link["href"]
        # filter for html pages with absolute path
        if 'http' in linkUrl and '.html' in linkUrl:
            print(linkUrl)
            tmpSoup = safeOpenParsePage(linkUrl)
            if tmpSoup is None:
                continue
            else:
                addPageWordCountsToDict(word_dict, tmpSoup)
                counter = counter + 1
        if counter > 9: break

    printTopWordsinDict(word_dict, 250)
```

```python
import requests
from bs4 import BeautifulSoup
import ssl

# Disable SSL warnings
requests.packages.urllib3.disable_warnings()

def safeOpenParsePage(url):
    """
    Fetches and parses a webpage safely, handling exceptions like SSL errors, timeouts, and 404 errors.
    """
    try:
        response = requests.get(url, timeout=5, verify=False)  # Disable SSL verification
        response.raise_for_status()
        return BeautifulSoup(response.text, 'html.parser')
    except requests.exceptions.RequestException as e:
        print(f"Failed to open {url}: {e}")
        return None

def addPageWordCountsToDict(word_dict, soup):
    """
    Extracts word counts from paragraph text and updates a dictionary.
    """
    paragraphs = soup.find_all('p')  # Extract all paragraph elements
    for paragraph in paragraphs:
        words = paragraph.get_text().lower().split()
        for word in words:
            word = word.strip(".,!?()[]{}:;\"'")  # Remove punctuation
            if word:
                word_dict[word] = word_dict.get(word, 0) + 1

def printTopWordsinDict(word_dict, num_words=250):
    """
    Prints the top N most frequent words from the dictionary.
    """
    sorted_words = sorted(word_dict.items(), key=lambda x: x[1], reverse=True)
    for word, count in sorted_words[:num_words]:
        print(f"{word}: {count}")

# Define the target page
targetPage = "https://en.wikipedia.org/wiki/Purdue_University"

# Open and parse the target page
soupPurdue = safeOpenParsePage(targetPage)

if soupPurdue:
    # Initialize the word count dictionary
    word_dict = {}

    # Count words on the Purdue page
    addPageWordCountsToDict(word_dict, soupPurdue)

    # Find all valid links on the page
    links = soupPurdue.find_all('a', href=True)

    # Add word counts from the first ten linked pages
    counter = 0
    for link in links:
        if counter >= 10:
            break

        linkUrl = link["href"]

        # Ensure it's an absolute URL (ignoring Wikipedia internal links)
        if linkUrl.startswith("http"):
            print(f"Processing: {linkUrl}")
            tmpSoup = safeOpenParsePage(linkUrl)
            if tmpSoup:
                addPageWordCountsToDict(word_dict, tmpSoup)
                counter += 1
            else:
                print(f"Skipping {linkUrl} due to an error.")

    # Print the top 250 words
    printTopWordsinDict(word_dict, 250)
```
