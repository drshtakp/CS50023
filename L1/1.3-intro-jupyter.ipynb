```python
class student:
    """A class representing a student."""
    def __init__(self,n,a):
        self.full_name = n
        self.age = a
    def get_age(self):
        return self.age
    def set_age(self,num):
        self.age = num

b = student("Bob", 21)
print('initial age', b.get_age())

b.set_age(18)
print('new age', b.get_age())

print(b.full_name, b.age)
```
